```python
import random
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
array = [random.randint(0, 100) for _ in range(6)]
print("Початковий масив:", array)
sorted_array = bubble_sort(array.copy())
print("Відсортований масив:", sorted_array)
```

```python
import time
import matplotlib.pyplot as plt
import random
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000]
insertion_times = []
bubble_times = []
for size in sizes:
    array = [random.randint(0, 3000) for _ in range(size)]
    start_time = time.time()
    insertion_sort(array.copy())
    end_time = time.time()
    insertion_times.append(end_time - start_time)
    start_time = time.time()
    bubble_sort(array.copy())
    end_time = time.time()
    bubble_times.append(end_time - start_time)
plt.plot(sizes, insertion_times, label='Insertion Sort')
plt.plot(sizes, bubble_times, label='Bubble Sort')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Sorting Time Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
